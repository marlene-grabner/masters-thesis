### 1. HPO symptoms dataset

```python
def parse_HPO(filename):
    #regex to search
    id_pattern = re.compile(r'id: HP:(\d+)')
    name_pattern = re.compile(r'name: (.+)')
    mesh_pattern = re.compile(r'xref: MSH:(\S+)')
    umls_pattern = re.compile(r'xref: UMLS:(\S+)')
    is_a_pattern = re.compile(r'is_a: HP:\d+ ! (.+)')
    
    #storage lists
    ids = []
    names = []
    mesh_ids = []
    umls_ids = []
    symptoms = []
    
    #open file
    with open(filename, 'r') as file:
        lines = file.readlines()
        #current variables
        current_id = None
        current_name = None
        current_mesh_id = []
        term_start = False  #flag to indicate the start of a term
        current_umls = []
        umls_found = False
        current_symptoms = []  #symptoms list (there is more than one symptom per term)
        mesh_found = False  #flag to indicate if a Mesh ID has been found for the current term
        
        #iterate through lines in entry, entry starts with [Term]
        for line in lines:
            #check if the line marks the start of a term entry
            if line.strip() == '[Term]':
                term_start = True
                if current_id is not None and current_symptoms:
                    # Append symptoms for the previous term
                    for symptom in current_symptoms:
                        for umls_id in current_umls if umls_found else [None]:
                            for mesh_id in current_mesh_id if mesh_found else [None]:
                                ids.append(current_id)
                                names.append(current_name)
                                mesh_ids.append(mesh_id)
                                umls_ids.append(umls_id)
                                symptoms.append(symptom)
                    current_symptoms, current_umls, current_mesh_id = [], [], []  # Reset symptoms, mesh id and umls for the new term
                    mesh_found, umls_found = False, False  # Reset mesh_found flag
                continue
            
            # If not start of a term, skip processing the line
            if not term_start:
                continue
            
            # Extract id, name, Mesh ID, and symptoms if present in the line
            id_match = id_pattern.match(line)
            if id_match:
                current_id = id_match.group(1)

            name_match = name_pattern.match(line)
            if name_match:
                current_name = name_match.group(1)

            mesh_match = mesh_pattern.match(line)
            if mesh_match:
                current_mesh_id.append(mesh_match.group(1))
                mesh_found = True  # Set mesh_found flag to True

            umls_match = umls_pattern.match(line)
            if umls_match:
                current_umls.append(umls_match.group(1))
                umls_found = True

            is_a_match = is_a_pattern.match(line)
            if is_a_match:
                current_symptoms.append(is_a_match.group(1))  # Store symptom in list
            
            # Check if empty line marks end of entry
            if line.strip() == '':
                term_start = False  # Reset start flag for next term
        
        # Append parsed values to lists (repeated for each symptom in current_symptoms)
        # to get individual lines per symptom
        if current_id is not None and current_symptoms:
            for symptom in current_symptoms:
                for single_umls in current_umls if umls_found else [None]:
                    for mesh_id in current_mesh_id if mesh_found else [None]:
                        ids.append(current_id)
                        names.append(current_name)
                        mesh_ids.append(mesh_id)
                        umls_ids.append(single_umls)
                        symptoms.append(symptom)

    # Make DataFrame
    df = pd.DataFrame({'ID': ids, 'Disorder': names, 'Mesh_ID': mesh_ids, 'UMLS': umls_ids, 'Symptom': symptoms})
    return df
```

### Synonyms would not be recognized as such
--> fuzzywuzzy matching to try and get higher similarity

```python
def replace_similar_names(df, threshold=80):
    from fuzzywuzzy import process
    symptom_map = {}
    count = 0
    x = 0
    y = 0

    for symptom in df['Symptom']:
        if symptom_map:
            best_match, score = process.extractOne(symptom, symptom_map.keys())
            if best_match and score >= threshold:
                symptom_map[symptom] = symptom_map[best_match]
                if symptom != best_match:
                    print(f"{symptom} is now called {best_match}")
                    x += 1
                if symptom == best_match:
                    count += 1
            else:
                y += 1
                symptom_map[symptom] = symptom
        else:
            y += 1
            symptom_map[symptom] = symptom

    df['Symptom'] = df['Symptom'].map(symptom_map)
    print(f"Newly added: {y}")
    print(f"Matched to the same: {count}")
    print(f"Actual new match: {x}")
    return df
```

```python
def split_genes(df):
    rows = []
    
    for idx, row in df.iterrows():
        if pd.isna(row['gene']):
            # Handle NaN: either skip or keep as is
            rows.append(row)
        else:
            genes = row['gene'].split(', ')
            for gene in genes:
                new_row = row.copy()
                new_row['gene'] = gene
                rows.append(new_row)
    
    new_df = pd.DataFrame(rows)
    return new_df
```

```python
#Putting genes assoicated with several diseases into their own rows
gene_merge  = split_genes(gene_merge_unp)
gene_merge
```
